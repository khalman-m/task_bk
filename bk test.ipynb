Каждый кандидат характеризуется следующими атрибутами:

```python
import numpy as np
import pandas as pd
```

```python
def generate_data(n, k_prob=0.05, k_min = 3, k_max=50, ctr_beta_a=1, ctr_beta_b=100, bid_logn_mu=1, bid_logn_sigma=1, bid_max = 5000):
    k = np.random.geometric(k_prob, n)
    k = k.clip(k_min, k_max)
    
    ctr = np.zeros((n, k_max))
    bid = np.zeros((n, k_max))
    for i in xrange(n):
        ctr[i,:k[i]] = np.random.beta(ctr_beta_a, ctr_beta_b, k[i])
        bid[i,:k[i]] = np.round(np.random.lognormal(bid_logn_mu, bid_logn_sigma, k[i])) # in cents
        bid[i, :k[i]] = bid[i,:k[i]].clip(min=0.01, max=bid_max)
        
    return k,ctr,bid.astype(np.int)

def save_data_into_file(filename, k, ctr, bid):
    n = len(k)
    with open(filename, 'w') as f:
        for i in xrange(n):
            for j in xrange(k[i] - 1):
                f.write("%0.6f,%d\t" % (ctr[i][j],bid[i][j]))
            f.write("%0.6f,%d\n" % (ctr[i][k[i] -1],bid[i][k[i] - 1]))
            
def load_data_from_file(filename, max_lines=100000):
    """
        according to problem discription
    """
    k = []
    ctr = []
    bid = []
    with open(filename, 'r') as f:
        for line,i in zip(f,xrange(max_lines)):
            print i
            candidates = [cand.strip() for cand in line.split('\t')]
            cands = [cur_cand.split(',') for cur_cand in candidates]    
            k.append(len(cands))
            ctr.append([float(cand[0]) for cand in cands])
            bid.append([int(cand[1]) for cand in cands])
    n = i
    m = np.max(k)
        
    # iterative conversion because lists are of different size
    # and converts to array of lists otherwise
    ctr_np = np.zeros((n, m))
    bid_np = np.zeros((n, m), dtype=np.int)
    for i in xrange(n):
        ctr_np[i][0:k[i]] = np.array(ctr[i])
        bid_np[i][0:k[i]] = np.array(bid[i])
        
        
    return np.array(k),ctr_np,bid_np
```

```python
def get_ads_to_show(k, ctr, bid, m, n=ctr.shape[0]):
    """
        Chooses ads to show according to ctr and bid for each query
        
        
        Usage:
            queries_to_use,ads_to_use,number_of_ads = get_ads_to_show(k, ctr, bid, m, n)
            
            queries_to_use - np.array (n x 1) of best queries for which ads are shown
            ads_to_use - np.array (n x m) of candidates (indexes) chosen for each query
            number_of_ads - np.array (n x 1) stores number of ads to show for each query 
            (can be less them m when k < m)
    """
    expected_profit = ctr * bid
    sorted_idx = np.argsort(-expected_profit, axis=1)
    total_profit = expected_profit.take(sorted_idx[:,:m]).sum(axis = 1)
    sorted_queries = np.argsort(-total_profit) 
    return sorted_queries[:n],sorted_idx[:,:m],k.clip(0, m)
```

```python
# Example:
k,ctr,bid = generate_data(10)
save_data_into_file('dataset.tsv', k, ctr, bid)
k,ctr,bid = load_data_from_file('dataset.tsv')
queries_to_use,ads_to_use,number_of_ads = get_ads_to_show(k, ctr, bid, m=3)
```
